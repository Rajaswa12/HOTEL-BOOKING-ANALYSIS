```python
#analysis of data and reservation 
cancelled_perc =da['is_canceled'].value_counts(normalize = True)
```

```python
cancelled_perc #percentage for percentage 
```

```python
plt.figure(figsize =(5,4))
plt.title('reservations status count')
plt.bar(['Not canceled','Canceled'],da['is_canceled'].value_counts(),color = 'purple',width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax1 = sns.countplot(x ='hotel',hue='is_canceled',data = da ,palette = 'Blues')
legend_labels, _ = ax1. get_legend_handles_labels()
plt.title('Reservation status in different hotels',size=18)#font size can be change with size 
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel = da[da['hotel'] == 'Resort Hotel'] #percentage cancellation
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = da[da['hotel'] == 'City Hotel'] #percentage cancellation
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# group by function use for grouping them accordingly #hypothesis
resort_hotel =resort_hotel.groupby('reservation_status_date')[['adr']].mean()

city_hotel =city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in city and Resort hotel',fontsize = 40)
plt.plot(resort_hotel.index, resort_hotel['adr'], label ='Resort Hotel',color='yellow')
plt.plot(city_hotel.index,city_hotel['adr'], label ='city Hotel',color='blue')
plt.legend(fontsize = 20)
plt.show()
```

```python
da['month'] = da['reservation_status_date'].dt.month #month conversion
#return month dt.month

da['month']
```

```python
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x ='month',hue='is_canceled',data = da ,palette = 'bright')
plt.title('Reservation status per month',size = 20)
plt.xlabel('months')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
plt.bar('month','adr', data = da[da['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
 cancelled_data = da[da['is_canceled'] ==1]
top_10_country = cancelled_data['country'].value_counts()[0:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_country,labels= top_10_country.index,autopct='%0.2f')
plt.title('top ten countries with reservation cancelled')
plt.show()
```

```python
#price effect in cancellation

cancelled_da_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()

cancelled_da_adr.reset_index(inplace = True)
cancelled_da_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data = da[da['is_canceled'] == 0]
not_cancelled_da_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_da_adr.reset_index(inplace = True)
not_cancelled_da_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize=(20,6))
plt.plot(not_cancelled_da_adr['reservation_status_date'],not_cancelled_da_adr['adr'],label='not cancelled')
plt.plot(cancelled_da_adr['reservation_status_date'],cancelled_da_adr['adr'],label='cancelled')
plt.title('average daily rate')
plt.legend()
```
